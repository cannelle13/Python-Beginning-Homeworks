b. Tabulate the following function: F(x)=2x^2+5, x∈[1,100] with step 1.

```python
x_values = np.arange(1, 101, 1)
fx_values = 2 * x_values**2 + 5

for x, fx in zip(x_values, fx_values):
    print(f"x = {x:<3d} | F(x) = {fx:<5d}")
```

c. Tabulate the following function: F(x)=e^−x, x∈[−10,10] with step 1.

```python
x_values = np.arange(-10, 11, 1)
fx_values = np.exp(-x_values)

for x, fx in zip(x_values, fx_values):
    print(f"x = {x:<3d} | F(x) = {fx:<10.6f}")
```

```python
planets = sns.load_dataset("planets")

# Трохи гралась з оформленням
# Зміна стилю та розміру графіка
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))

# Заголовок та підписи
plt.title("Distribution of Planet Mass")
plt.xlabel("Mass")
plt.ylabel("Frequency")

# Збільшення розміру шрифту
sns.set_context("notebook", font_scale=1.2)

# Гістограма з більшою кількістю бінів та зеленим кольором
sns.histplot(planets['mass'], bins=20, color='green')

# Додавання граничної лінії для середнього значення
mean_mass = planets['mass'].mean()
plt.axvline(mean_mass, color='red', linestyle='dashed', linewidth=2, label='Mean Mass')

# Показати графік
plt.show()
```

```python
sns.kdeplot(planets['mass'])
```

```python
sns.ecdfplot(planets['mass'])
```

```python
# Спроба обʼєднати в три як в лекції
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

sns.histplot(planets['mass'], ax=axs[0])
axs[0].set_title("Histogram")

sns.kdeplot(planets['mass'], ax=axs[1])
axs[1].set_title("KDE Plot")

sns.ecdfplot(planets['mass'], ax=axs[2])
axs[2].set_title("ECDF Plot")

plt.tight_layout()
plt.show()
```

```python
sns.displot(planets, x='year', hue='method', kind="hist", fill=True)
```

```python
sns.regplot(x='year', y='mass', data=planets)
```

```python
sns.catplot(x='year', y='method', data=planets, kind='strip')
```

```python
sns.catplot(y="method", x="year", kind="strip", data=planets, s=10)
```
